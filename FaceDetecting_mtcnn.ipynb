```python
from mtcnn.mtcnn import MTCNN
import cv2 
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
```

```python
img = cv2.imread('./UFC_test.jpeg')
IMG = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(IMG)
plt.axis('off')
```

```python
detector = MTCNN()
```

```python
def detectface(img):
    plt.imshow(img)
    ax = plt.gca()
    faces = detector.detect_faces(img)
    # print bounding box for each detected face
    for face in faces:
        x, y, w, h = face['box']
        rect = Rectangle((x, y), w, h, fill = False, color = 'red')
        ax.add_patch(rect)
        print(face)
    plt.show()
```

```python
detectface(IMG)
```

```python
def facelist(img):
    faces = detector.detect_faces(img)
    # print bounding box for each detected face
    for face in faces:
        print(face)
        
    return faces

def output_faces(img, facelist):
    for i in range(len(facelist)):
        x, y, w, h = facelist[i]['box']
        xn, yn = x + w, y + h
        plt.subplot(1, len(facelist), i+1)
        plt.axis('off')
        plt.imshow(img[y:yn, x:xn])
    plt.show
```

```python
FL = facelist(IMG)
```

```python
output_faces(IMG, FL)
```
