```python
import cv2
from matplotlib import pyplot as plt
```

```python
detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('./UFC_test.jpeg')
IMG = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(IMG)
plt.axis('off')
```

```python
def detectface(img):
    bboxes = detector.detectMultiScale(img)
    # print bounding box for each detected face
    for box in bboxes:
        x, y, w, h = box
        xn, yn = x + w, y + h
        cv2.rectangle(img, (x, y), (xn, yn), (255,0,0), 1)
        
    return img
```

```python
plt.imshow(detectface(IMG))
plt.axis('off')
```
